```python
import re
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f=open("md.txt",encoding="utf-8")
f
```

```python
data=f.read()
data
```

```python
type(data)
```

```python
print(data)
```

```python
pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages=re.split(pattern,data)
messages
```

```python
type(messages)
```

```python
messages=messages[1:]
messages
```

```python
dates=re.findall(pattern,data)
```

```python
dates
```

```python
df=pd.DataFrame({"user_message":messages,"message_date":dates})
df
```

```python
df.shape
```

```python
df.info()
```

```python
df["message_date"]=pd.to_datetime(df["message_date"],format="%d/%m/%y, %H:%M - ")
```

```python
df.dtypes
```

```python
df.head(10)
```

```python
df["year"]=df["message_date"].dt.year
df["month"]=df["message_date"].dt.month_name()
df["day"]=df["message_date"].dt.day
df["hour"]=df["message_date"].dt.hour
df["hour"]=df["message_date"].dt.month
df["day_name"]=df["message_date"].dt.day_name()
df["Date"]=df["message_date"].dt.date
df
```

```python
users=[]
message=[]
for i in df["user_message"]:
    print(i)
    if "changed tha subject" not in i:
        li=i.split(": ",maxsplit=1)
    else:
        li=[]
        li.append(i)
        
    if len(li)>=2:
        users.append(li[0])
        message.append(li[1])
    else:
        users.append("group notification")
        message.append(li[0])
df["user"]=users
df["message"]=messages
df
```

```python
df.drop(columns=["user_message"],inplace=True)
```

##  find most frequent chating words.

```python
df1=(df[df["user"]!="group notification"])
df1
```

```python
words=[]
for i in df1["message"]:
    li=i.split()
    words.extend(li)
words
```

```python
user_massage=df1[df1["message"]!='<Media omitted>\n']
words=[]
for i in user_massage["message"]:
    li=i.split()
    words.extend(li)
words
```

```python
from urlextract import URLExtract
```

```python
extractor=URLExtract()
word=[]
for i in words:
    if len(extractor.find_urls(i))==0:
        word.append(i)
word
```

```python
len(word)
```

```python
wordss=[]
for i in word:
    if ("<Media"!=i) and ("omitted>"!=i):
        wordss.append(i)
wordss
```
